```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ConstructionTimeSeriesDataV2.csv')  
```

```python
dates=[i for i in range(2002,2015)]
fig,ax = plt.subplots()
ax.plot(df['Month-Year'],df['Total Construction'], label='Total')
ax.plot(df['Month-Year'],df['Private Construction'],color='k',label='Private')
ax.plot(df['Month-Year'],df['Public Construction'],color='g',label='Public')
fig.suptitle('Spending on Construction during the Recession (2002-2014)',fontsize=14)   # Graph title
ax.xaxis.set_label_text('Date')  # x-axis caption
ax.yaxis.set_label_text('Spending ($)')       # y-axis caption
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True)  # Turn top x axis tick marks off 
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True) 
ax.axvspan(65,110,ymin=0,ymax = 100000,facecolor='k',alpha=0.1)
ax.axvline(x=65,linewidth=1,color='k',linestyle='--')
ax.axvline(x=110,linewidth=1,color='k',linestyle='--')
ax.text(88,12800,'During the Recession, Private spending\nwent down, while Public grew!',fontsize=11,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')
ax.text(75.5,115000,'Start of Recession',fontsize=11,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')
ax.legend(loc=1)
ax.set_ylim(0,max(df['Total Construction'])+.1*max(df['Total Construction']))
ax.set_xticks(df['Month-Year'][::12])
ax.set_xticklabels(dates)
fig.set_size_inches(12,6)
plt.show()
```
